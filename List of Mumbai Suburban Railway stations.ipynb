```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://en.wikipedia.org/wiki/List_of_Mumbai_Suburban_Railway_stations"
response = requests.get(url)
response
```

```python
soup = BeautifulSoup(response.content, "html.parser")
soup
```

```python

td_elements = soup.find_all("td")
td_elements    # td_elements
```

```python
# Extract text from each td element
list_general=[]
for td in td_elements:
    text = td.get_text()
    list_general.append(text)
list_general        # list_general
```

```python
search_first_row_text="Airoli"
search_last_row_text="Targhar"
print(search_first_row_text)
print(search_last_row_text)
```

```python
search_index_first_row=list_general.index(search_first_row_text)
search_index_last_row=list_general.index(search_last_row_text)

print("The index of the'Airoli'in above list is: ",search_index_first_row )
print("The index of the'Targhar' in the above list is: ",search_index_last_row)
```

```python
new_list=[]
for i in range(search_index_first_row-1,search_index_last_row+6):
    new_list.append(list_general[i])
print(new_list)
```

```python
name=[]
for i in range(1,len(new_list),8):
    name.append(new_list[i])
print(name)
```

```python
len(name)
```

```python
name_marathi=[]
for i in range(2,len(new_list),8):
    name_marathi.append(new_list[i])
print(name_marathi)
```

```python
len(name_marathi)
```

```python
R_line=[]
for i in range(4,len(new_list),8):
    R_line.append(new_list[i])
print(R_line)
```

```python
len(R_line)
```

```python
fast_train_halt=[]
for i in range(5,len(new_list),8):
    fast_train_stop.append(new_list[i])
print(fast_train_stop)
```

```python
len(fast_train_stop)
```

```python
long_distance=[]
for i in range(6,len(new_list),8):
    long_distance.append(new_list[i])
print(long_distance)
```

```python
len(long_distance)
```

```python
import pandas as pd

df=pd.DataFrame({"Railway_station_Name":name,"Name_in_marathi":name_marathi,"Railway_line":R_line,"Fast_Train_Halt":fast_train_stop,"Long_Distance":long_distance})
df
```

```python
df.to_csv("List of Mumbai Suburban Railway stations.csv")
```
